```python
from keras.models import Model, model_from_json, load_model

import matplotlib.pyplot as plt
import scipy.misc
import numpy as np
import os

%matplotlib inline
```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "" # do not use the GPU
```

```python
nb_classes = 5
# input image dimensions
img_rows = img_cols = 512
layers = 3  

#Load a existing model architecture
filepath = "Json2016-11-17.txt"
with open(filepath, 'r') as myfile:
    json_string=myfile.read()

model = model_from_json(json_string)
#load Weights 
model.load_weights("2016-11-17_eph_126_acc_0.58.hdf5")


testPath = "retina_images"
X_test = np.load(os.path.join(testPath, "images_1.npy"))  
y_test = np.load(os.path.join(testPath, "labels_1.npy")) 

for k,img in enumerate(X_test[0:6]):
    prob = model.predict_proba(np.expand_dims(img,axis=0), batch_size=1, verbose=0)
    cl = np.argmax(prob)
    print(prob)
    print('class:',cl)
    implot = np.rollaxis(img,0,3)
    plt.figure(figsize=(6,6))
    plt.imshow(implot[:,:,::-1])
    plt.show()
    scipy.misc.imsave('image_'+str(k)+'_class_'+str(cl)+'.jpg', implot)
```
